# One subject and block

```python
sid = '73T01'
cid = 'block1'
```

```python
# Segment raw data into different epoch-types
epos_dict = get_all_epo_objects(raw_rec, sid=sid, cid=cid, save=save, load=load, verbose=False, test=test)
epos_dict
```

# Multiple subjects and their blocks at once

```python
sids = ['73T01']
```

```python
for sid in sids:
    print(f"{sid = }")
    cids = get_sid_cids(sid, test=test)
    for cid in cids:
        print(f"{cid = }")
        
        ## 1. Extract epochs
        
        # Load clean continuous (raw) data
        raw_rec = get_raw_to_epoch(sid, cid)
        
        # Segment raw data into different epoch-types
        epos_dict = get_all_epo_objects(raw_rec, sid=sid, cid=cid, save=save, load=load, verbose=False, test=test)
        
        ## 2. Visualize extracted epochs
        
        # Plot evoked response 
        _ = plot_evk_by_grp(epos_dict, sid=sid, cid=cid, show=show, save=save)
        
        # Compute and plot PSD
        psd_by_rec = compute_psd_by_key(epos_dict)
        _ = plot_psd_avg_by_grp(psd_by_rec, sid, cid, show=show, save=save)
```
