```python
from visualization.vis_eeg import plot_evk_by_cat, plot_psd_avg_by_cat
from preprocessing.extract_eeg  import get_raw_to_epoch, get_all_epo_objects
from processing.process_eeg import compute_psd_by_key
from utils.gen_utils import get_sid_cids

%load_ext autoreload
%autoreload 2
```

```python
test = False
show = False
save = False
load = False
```

# One participant and block at a time

## 1. Extract epochs 
### Load clean raw data

```python
pid = '73T01'
cid = 'block1'
```

```python
raw_rec = get_raw_to_epoch(pid, cid)
```

```python
epos_dict = get_all_epo_objects(raw_rec, pid=pid, cid=cid, save=save, load=load, verbose=False, test=test)
epos_dict
```

## 2. Visualize 

### Evoked

```python
# Plot evoked response (average across epochs)
%matplotlib inline
_ = plot_evk_by_cat(epos_dict, pid=pid, cid=cid, show=show, save=save)
```

### PSD

```python
psd_by_rec = compute_psd_by_key(epos_dict)
_ = plot_psd_avg_by_cat(psd_by_rec, pid, cid, show=show, save=save)
```

# Multiple participant/blocks at once

```python
pids = ['73T01']
```

```python
for pid in pids:
    print(f"{pid = }")
    cids = get_sid_cids(pid, test=test)
    for cid in cids:
        print(f"{cid = }")
        
        raw_rec = get_raw_to_epoch(pid, cid)
        epos_dict = get_all_epo_objects(raw_rec, pid=pid, cid=cid, save=save, load=load, verbose=False, test=test)
        
        ## 2. Visualize 
        
        # Plot evoked response (average across epochs)
        _ = plot_evk_by_cat(epos_dict, pid=pid, cid=cid, show=show, save=save)
        
        # Plot PSD
        psd_by_rec = compute_psd_by_key(epos_dict)
        _ = plot_psd_avg_by_cat(psd_by_rec, pid, cid, show=show, save=save)
```
