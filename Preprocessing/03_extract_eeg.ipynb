# Multiple subject/blocks at once

```python
segmented_epochs_options = [True]
```

```python
for sid, cids in cids_by_sid.items():
    print(f"{sid = }")
    for cid in cids:
        print(f"{cid = }")
        
        for seg_opt in segmented_epochs_options:

            raw_rec = get_raw_to_epoch(sid, cid)
            epos_dict = get_all_epo_objects(raw_rec, sid=sid, cid=cid, save=save, load=load, verbose=False, test=test, segmented_epochs=seg_opt)
            
            ## 2. Visualize 
            
            # Plot evoked response (average across epochs)
            plot_evk_by_cat(epos_dict, sid=sid, cid=cid, show=show, save=save, segmented_epochs=seg_opt)
            
            # Plot PSD
            psd_by_rec = compute_psd_by_key(epos_dict)
            plot_psd_avg_by_cat(psd_by_rec, sid, cid, show=show, save=save, segmented_epochs=seg_opt)
```
