```python
from preprocessing.behavior_to_eeg import get_times_retrieval_phases, get_trace_df, get_retrieval_df, extract_behavioral_events, define_eeg_epochs 
from visualization.vis_eeg import plot_schematic_epo_def

%load_ext autoreload
%autoreload 2
```

```python
test = False
show = True
save = True
```

# One subject and block at a time

## 1. Extract epochs 
### Load clean raw data

```python
segment_epochs = True
```

```python
sid = '04'
```

```python
retr_times = get_times_retrieval_phases(sid)
retr_times
```

```python
trace_df = get_trace_df(sid)
# trace_df
```

```python
retrieval_df = get_retrieval_df(sid)
# retrieval_df
```

```python
beh_events_df = extract_behavioral_events(sid, retr_times, retrieval_df, trace_df, save=save)
# beh_events_df
```

```python
eeg_epochs_df = define_eeg_epochs(beh_events_df, sid, save=save, segment_epochs=segment_epochs)
# eeg_epochs_df
```

```python
counter = 0
for block in beh_events_df['RetrievalBlock'].unique():
    for trial in beh_events_df['Trial'].unique():
        plot_schematic_epo_def(beh_events_df, eeg_epochs_df, block_n=block, trial_n=trial, show=show, save=save, sid=sid, segmented_epochs=segment_epochs)
        counter += 1
        if test and counter > 0:
            break
    if test and counter > 0:
            break
```

# Multiple subject/blocks at once

```python
sids = ['02', '03', '04', '05']
```

```python
segment_epochs_options = [False, True]
```

```python
for sid in sids:
    for seg_opt in segment_epochs_options:
        block_times = get_times_retrieval_phases(sid)
        df_trace = get_trace_df(sid)
        retrieval_df = get_retrieval_df(sid)
        beh_events_df = extract_behavioral_events(sid, block_times, retrieval_df, df_trace, save=save)
        eeg_epochs_df = define_eeg_epochs(beh_events_df, sid, save=save, segment_epochs=seg_opt)
        for block in beh_events_df['RetrievalBlock'].unique():
            for trial in beh_events_df['Trial'].unique():
                plot_schematic_epo_def(beh_events_df, eeg_epochs_df, block_n=block, trial_n=trial, show=show, save=save, sid=sid, segmented_epochs=seg_opt)
```
